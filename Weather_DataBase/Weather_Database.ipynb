# Retrieve Weather Data

```python
# Import Dependencies

import pandas as pd
import numpy as np
from citipy import citipy
import requests
from config import weather_api_key
```

```python
# 3. Create a new set of 2,000 random latitudes and longitudes.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
coordinates = list(lat_lngs)
```

```python
# Check the list of coordinates
len(coordinates)
```

```python
# 4. Get the nearest city using citipy module.

# Create an empty list for cities
cities = []

# Identify nearest city for each coordinate
for coordinate in coordinates: 
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # Add city to the list if not on the list
    if city not in cities: 
        cities.append(city)
        
# Confirm cities list
len(cities)
```

```python
# 5. Perform an API call with the OpenWeatherMap.

# Build base URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# Create an empty list for the weather data
city_data = []

# Print to mark the beginning of the logging
print("Retrieving Data from OpenWeatherMap API    ")
print("-------------------------------------------")

# Initialize counters
record_count = 1
set_count = 1

# Loop through the cities with the enumerate() method
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
    
    # Log the URL, record, and set numbers and the city. 
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Increment record_count by 1
    record_count += 1
    
    # Run an API request for each city with try-except block
    try:
        # Parse the JSON and retrieve data. 
        city_weather = requests.get(city_url).json()
    
        # 6. Retrieve the needed data: Lat & Lng, Max Temp, % Humidity, % Cloudiness, Wind Speed, Weather Description
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]
    
        # Append the city information into city_data list. 
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Current Description": city_description})
    
    # Skip city if there is an error.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Check the list
len(city_data)
```

```python
# 7a. Add the data to a new DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Reorder the columns
new_col_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[new_col_order]

# 7b. Confirm DataFrame complies with requirements
city_data_df.head(10)
```

```python
# Check the DataFrame for null values
city_data_df.isnull().values.any()
```

```python
# 8. Export the DataFrame as a CSV File and save
output_data_file = "WeatherPy_Database.csv"
city_data_df.to_csv(output_data_file)
```
